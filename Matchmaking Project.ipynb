```python
import pandas as pd
import numpy as np
```

```python
People = []
passions = []
```

```python
gender_options = ['Male','Female','Other']
hair_color_options = ['Brown','Blond','Black','Blue','White','Red','Mixed','Other']
hair_texture_options = ['Straight','Wavy','Curly','Frizzy','Other']
eye_color_options = ['Brown','Blue','Red','Amber','Gray','Green','Hazel','Mixed','Other']

gender_seeking_options = ['Male','Female','Other','Unimportant']
hair_color_seeking_options = ['Brown','Blond','Black','Blue','White','Red','Mixed','Other','Unimportant']
hair_texture_seeking_options = ['Straight','Wavy','Curly','Frizzy','Other','Unimportant']
eye_color_seeking_options = ['Brown','Blue','Red','Amber','Gray','Green','Hazel','Mixed','Other','Unimportant']

columns = ['Characteristic_age','Characteristic_Eye_Color','Characterisic_Gender','Characteristic_Hair_Texture','Characteristic_Hair_Color','Characteristic_Height','Characteristic_Passions','Name','Number','Seeking_Age_Max','Seeking_Age_Min','Seeking_Eye_Color','Seeking_Gender','Seeking_Hair_Color','Seeking_Hair_Texture','Seeking_Height']
```

```python
name = input("Enter your full name: ")
number = input("Enter your contact phone number: ")

while True:
    age = input('Enter your age: ')
    try:
        if int(age) >= 16 and int(age) < 100:
            break
        else:
            print("\nNot A Valid Input, Please Enter a Number Between 16 and 99.\n")
    except ValueError:
        print("\nNot a Valid Input, Please Enter a Number Between 16 and 99.\n")

while True:
    gender = input('Enter your gender: ')
    if gender.capitalize() in gender_options:
        break
    else:
        print("\nNot a Valid Input, Please Enter 'Male', 'Female', or 'Other'.\n")

while True:
    hair_color = input('Enter your hair color: ')
    if hair_color.capitalize() in hair_color_options:
        break
    else:
        print("\nNot a Valid Input, Please Enter a Value From The Following Options:\n",hair_color_options)

while True:
    hair_texture = input('Enter your hair texture: ')
    if hair_texture.capitalize() in hair_texture_options:
        break
    else:
        print("\nNot a Valid Input, Please Enter a Value From The Following Options:\n",hair_texture_options)

while True:
    eye_color = input('Enter your eye color: ')
    if eye_color.capitalize() in eye_color_options:
        break
    else:
        print("\nNot a Valid Input, Please Enter a Value From The Following Options:\n",eye_color_options)

while True:
    height = input('Enter your height (in inches, 5ft = 60in): ')
    try:
        if int(height) >= 48 and int(age) < 96:
            break
        else:
            print("\nNot a Valid Input, Please Enter a Number of Inches Between 48 and 96.\n")
    except ValueError:
        print("\nNot a Valid Input, Please Enter a Number of Inches Between 48 and 96.\n")

while True:
    passion = input("Enter your passions/interests/hobbies, 1 by 1. Type 'Done' when you are finished: ")
    if passion.capitalize() == 'Done':
        print("\nFinished!\n")
        break
    else:
        passions.append(passion.capitalize())

while True:
    age_min_seeking = input("Enter your partners desired MINumum age, or type 'Unimportant': ") 
    if age_min_seeking.capitalize() != 'Unimportant':
        try:
            if int(age_min_seeking) >= 16 and int(age_min_seeking) < 100:
                break
            else:
                print("\nNot A Valid Input, Please Enter a Number Between 16 and 99.\n")
        except ValueError:
            print("\nNot a Valid Input, Please Enter a Number Between 16 and 99.\n")
    else:
        break

while True:
    age_max_seeking = input("Enter your partners desired MAXimum age, or type 'Unimportant': ")
    if age_max_seeking.capitalize() != 'Unimportant':
        try:
            if int(age_max_seeking) >= 16 and int(age_max_seeking) < 100:
                if (int(age_max_seeking) - int(age_min_seeking)) > -1:
                    break
                else:
                    print("\nNot a Valid Input, Please Make Sure This Value Is Equal to Or Larger Than Your Partners Desired Minimum Age.\n")
            else:
                print("\nNot A Valid Input, Please Enter a Number Between 16 and 99.\n")
        except ValueError:
            print("\nNot A Valid Input, Please Enter a Number Between 16 and 99.\n")
    else:
        break

while True:
    gender_seeking = input("Enter your partners desired gender, or type 'Unimportant': ")
    if gender_seeking.capitalize() == "Unimportant":
        break
    else:
        if gender_seeking.capitalize() in gender_seeking_options:
            break
        else:
            print("\nNot a Valid Input, Please Enter 'Male', 'Female', 'Other', or 'Unimportant'.\n")

while True:
    hair_color_seeking = input("Enter your partners desired hair color, or type 'Unimportant': ")
    if hair_color_seeking.capitalize() in hair_color_seeking_options:
        break
    else:
        print("\nNot a Valid Input, Please Enter a Value From The Following Options:\n",hair_color_seeking_options)

while True:
    hair_texture_seeking = input("Enter your partners desired hair texture, or type 'Unimportant': ")
    if hair_texture_seeking.capitalize() in hair_texture_seeking_options:
        break
    else:
        print("\nNot a Valid Input, Please Enter a Value From The Following Options:\n",hair_texture_seeking_options)

while True:
    eye_color_seeking = input("Enter your partners desired eye color, or type 'Unimportant': ")
    if eye_color_seeking.capitalize() in eye_color_seeking_options:
        break
    else:
        print("\nNot a Valid Input, Please Enter a Value From The Following Options:\n",eye_color_seeking_options)

while True:
    height_seeking = input("Enter your partners desired height (in inches, 5ft = 60in), or type 'Unimportant': ")
    if height_seeking.capitalize() == 'Unimportant':
        break
    else:
        try:
            if int(height_seeking) >= 48 and int(height_seeking) < 96:
                break
            else:
                print("\nNot a Valid Input, Please Enter a Number of Inches Between 48 and 96, or 'Unimportant'.\n")
        except ValueError:
            print("\nNot a Valid Input, Please Enter a Number of Inches Between 48 and 96, or 'Unimportant'.\n")

People.append({
    'Name': name,
    'Number': number,

    'Characteristic_Age': age.capitalize(),
    'Characteristic_Gender': gender.capitalize(),
    'Characteristic_Hair_color': hair_color.capitalize(),
    'Characteristic_Hair_Texture': hair_texture.capitalize(),
    'Characteristic_Eye_Color': eye_color.capitalize(),
    'Characteristic_Height': height.capitalize(),
    'Characteristic_Passions': passions,

    'Seeking_Age_Min': age_min_seeking.capitalize(),
    'Seeking_Age_Max': age_max_seeking.capitalize(),
    'Seeking_Gender': gender_seeking.capitalize(),
    'Seeking_Hair_Color': hair_color_seeking.capitalize(),
    'Seeking_Hair_Texture': hair_texture_seeking.capitalize(),
    'Seeking_Eye_Color': eye_color_seeking.capitalize(),
    'Seeking_Height': height_seeking.capitalize()
})

pd.DataFrame(People)
df = pd.DataFrame(People)
df.to_csv('People.csv')

print("\nInformation Successfully Inputted! \n")
```
